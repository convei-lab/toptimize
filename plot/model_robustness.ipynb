```python
import matplotlib.pyplot as plt
import numpy as np

RA_x=['0.5', '1.0', '1.5', '2.0']
PA_x=['0.05','0.1','0.15','0.2']

RA_GCN = [67.63, 58.52, 52.86, 46.08]
RA_GCN_TAD = [71.51, 63.90, 58.06, 49.06]
PA_GCN = [71.23, 56.24, 42.74, 33.54]
PA_GCN_TAD = [77.82, 60.32, 38.58, 26.11]
fig=plt.figure()
fig.show()
ax=fig.add_subplot(111)

ax.plot(RA_x,RA_GCN,c='r',marker="o",markersize=5, ls='-',label='RA_GCN_A')
ax.plot(RA_x,RA_GCN_TAD,c='g',marker="s",ls='-',label='RA_GCN_perturbA')
ax.plot(PA_x,PA_GCN,c='b',marker="^",ls='-',label='PA_GCN_oriA')
ax.plot(PA_x,PA_GCN_TAD,c='b',marker="^",ls='-',label='PA_GCN_perturbA')
ax.set_xlabel("Ratio of Perturbation")
ax.set_ylabel("Accuarcy")
plt.legend(loc=1)
plt.draw()
plt.savefig('Topology.pdf')
```
